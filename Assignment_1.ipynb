```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score
```

```python
df = pd.read_csv("./Downloads/Heart.csv")
```

```python
print("Shape of data:", df.shape)
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
print("\nData types:\n", df.dtypes)
```

```python
print("\nZero count in each column:\n", (df == 0).sum())
```

```python
print("\nMean age of patients:", df["Age"].mean())
```

```python
data = df[["Age", "Sex", "RestBP", "Chol", "MaxHR"]]
```

```python
train, test = train_test_split(data, test_size=0.25, random_state=42)
print("\nTraining set shape:", train.shape)
print("Test set shape:", test.shape)
```

```python
actual = np.concatenate((np.ones(100), np.zeros(400)))
predicted = np.concatenate((np.ones(120), np.zeros(380)))
```

```python
ConfusionMatrixDisplay.from_predictions(actual, predicted)
```

```python
print("\nClassification Report:\n", classification_report(actual, predicted))
print("Accuracy Score:", accuracy_score(actual, predicted))
```

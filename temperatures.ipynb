```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = pd.read_csv("./Downloads/temperatures.csv")
```

```python
print("\nDataset Information:")
print(data.info())
```

```python
print("\nStatistical Summary:")
print(data.describe())
```

```python
 print("\nMissing Values per Column:")
print(data.isnull().sum())
```

```python
print("\nZero Values per Column:")
print((data == 0).sum())
```

```python
X = data[["YEAR"]]
y = data["ANNUAL"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title("Temperature vs Year (Linear Regression)")
plt.xlabel("Year")
plt.ylabel("Annual Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.regplot(x="YEAR", y="ANNUAL", data=data, ci=None, color='red', line_kws={"lw":2})
plt.title("Simple Linear Regression using Seaborn")
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("\nModel Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R-Square (R²): {r2:.4f}")
```
